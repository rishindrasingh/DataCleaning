```python
import pandas as pd

import numpy as np


df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df.index.name='Index'

df
```

```python
flightnumber=[]

flightnumber=df["FlightNumber"]

l=len(flightnumber)
i=0
while(l>0):
    if np.isnan(df.FlightNumber[i]):
        df.FlightNumber[i]=df.FlightNumber[i-1]+10
    i+=1
    l-=1

df.FlightNumber = df.FlightNumber.astype(int)
df
        
```

```python
df_temp = pd.DataFrame(df.From_To.str.split('_',1).tolist(),
                                   columns = ['From','To'])

df_temp
```

```python
df_temp["From"]=df_temp.From.str.capitalize()

df_temp["To"]=df_temp.To.str.capitalize()

df_temp
```

```python
df["From"]=df_temp["From"]
df["To"]=df_temp["To"]

del df["From_To"]

df
```

```python
d2 = df['RecentDelays'].apply(pd.Series)

for col in d2:
    df["Delay%d" % (col+1)] = d2[col]

    
del df["RecentDelays"]
df.head()
```
